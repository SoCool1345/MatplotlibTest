```python
x = np.linspace(-3,3,50)
y1 = 2*x+1
y2 = x**2
plt.figure()
plt.plot(x,y1)
plt.figure(num=3,figsize=(8,5))
plt.plot(x,y1,linewidth=1.0,linestyle='--')
plt.plot(x,y2)
#设置展示坐标范围
plt.xlim((-1,2))
plt.ylim((-2,3))
plt.xlabel('i am x')
new_ticks = np.linspace(-1,2,5)
#设置坐标轴刻度
plt.xticks(new_ticks)
plt.yticks([-2,-1,1.5],['$really\ bad$','normal','good'])
ax = plt.gca()
#ax.spines['right'].set_color('none')
#ax.spines['top'].set_color('none')
#设置坐标轴位置
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.spines['bottom'].set_position(('data',0))
ax.spines['left'].set_position(('data',0))
#添加注解
x0 = 0.5
y0 = 2*x0+1
#方法1
plt.scatter(x0,y0,s=50,color='b') #散点图
plt.plot([x0,x0],[y0,0],'k--',lw = 2.5) # k black
plt.annotate(r'$2x+1=%s$' % y0,xy = (x0,y0),xycoords='data',
             xytext=(+30,-30),textcoords='offset points',fontsize=16,
             arrowprops=dict(arrowstyle='->',connectionstyle='arc3,rad=.2'))

#方法2
plt.text(-1,1,r'$this\ is\ text\ \mu\ \alpha_t\ \sigma_i$',fontdict={'size':16,'color':'r'})

#坐标轴能见度
for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(12)
    label.set_bbox(dict(facecolor='white',edgecolor='None',alpha=0.2))
```

```python
#散点图
n = 1024
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)
T = np.arctan2(Y,X)
print(T)
plt.scatter(X,Y,s=75,c=T,alpha=0.5)
plt.xlim((-1.5,1.5))
plt.ylim((-1.5,1.5))
plt.xticks(())
plt.yticks(())
```

```python
#条形图
n = 12
X = np.arange(n)
Y1 = (1-X/float(n))*np.random.uniform(0.5,1.0,n)
Y2 = (1-X/float(n))*np.random.uniform(0.5,1.0,n)
plt.bar(X,+Y1,facecolor='#9999ff',edgecolor='white')
plt.bar(X,-Y1,edgecolor='white')
print(Y1,Y2)
for x,y in zip(X,Y1):
    plt.text(x,y+0.05,'%.2f'%y,ha='center',va='bottom') #ha horizontal alignment 横向对齐, va 纵向对齐
for x,y in zip(X,Y2):
    plt.text(x,-y-0.05,'-%.2f'%y,ha='center',va='bottom')
plt.xlim((-1.5,n))
plt.ylim((-1.25,1.25))
plt.xticks(())
plt.yticks(())
```

```python
#等高线图
def f(x,y):
    return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2)

n = 256
x = np.linspace(-3,3,n)
y = np.linspace(-3,3,n)
X,Y = np.meshgrid(x,y)
print(X,Y,f(X,Y))
plt.contourf(X,Y,f(X,Y),20,alpha = 0.75,cmap=plt.cm.hot) #色彩图
C = plt.contour(X,Y,f(X,Y),20,colors='black',linewidths=.5) #等高线
plt.clabel(C,inline=True,fontsize=10)
```

```python
fig = plt.figure()
ax = Axes3D(fig)
X = np.arange(-4,4,0.25)
Y = np.arange(-4,4,0.25)
X,Y = np.meshgrid(X,Y)
R = np.sqrt(X**2+Y**2)
Z = np.sin(R)
ax.plot_surface(X,Y,Z,rstride=1,cstride=1,cmap=plt.get_cmap('rainbow'),edgecolor='black') #rstride行跨度
ax.contourf(X,Y,Z,zdir='z',offset=-2,cmap='rainbow') #offset 等高线图偏移量
ax.set_zlim(-2,2)
```

```python
#动画
X = np.arange(0, 10, 0.01)  # X shape： (N,)
Ys = [np.sin(X + k/10.0) for k in range(100)]  # Ys shape： (k, N)

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(8, 4))

def animate(i):
    axes[0].cla()
    axes[0].plot(X, Ys[i])
    axes[0].set_title(f'y = sin(x + {i}/10)')

    axes[1].cla()
    axes[1].hist(Ys[i], bins=50, orientation='horizontal')

ani = animation.FuncAnimation(fig, animate, frames=50, interval=50)
ani.save('matplotlib-animation-hist.gif')
plt.show()
```
